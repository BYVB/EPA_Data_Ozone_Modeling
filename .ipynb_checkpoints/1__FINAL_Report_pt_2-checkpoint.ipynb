## Prepping the data for regressions...

I saw in the air quality results that my yearly county rolling mean values are not quite normally distributed. Though close. In light of this fact, I tried a battery of logistic and non parametric models to see which yielded results. I also turned this into a classification question of predicting "good" air quality scores or "bad" ones (aka over 100 AQI).

```python
le = LabelEncoder()
```

```python
TRI_14_sm['CLEAR_AIR_ACT_CHEMICAL'] = le.fit_transform(TRI_14_sm['CLEAR_AIR_ACT_CHEMICAL'])
TRI_14_sm['CARCINOGEN'] = le.fit_transform(TRI_14_sm['CARCINOGEN'])
TRI_14_sm['UNIT_OF_MEASURE'] = le.fit_transform(TRI_14_sm['UNIT_OF_MEASURE'])
```

```python
TRI_14_sm.head()
TRI_14_sm['COUNTY'] = TRI_14_sm['COUNTY'].apply(lambda x: x.lower())
```

Here I am merging in my y values from the AQI score data I pickled in. 

```python
cty_yr_df = pd.DataFrame(all_cty_yr_dict).T
cty_yr_df = cty_yr_df.reset_index()
cty_yr_df.columns = ['County', 2013, 2014, 2015]
cty_yr_df = cty_yr_df.dropna()
cty_yr_df['County'] = cty_yr_df['County'].apply(lambda x: x.lower())
```

## Results and takeaways

My research suggests that there is the most predictive signal of air quality in the features related to the industry sector of a polluting plant. 

NAICS 334 (Computer and Electronic Manufacturing) accounts for almost a third of this feature's predictive power, with Petroleum (4247) and Hazardous Waste(562) directly behind. 

While these specific industries dominate the industry specific model, it's worth noting that this model's accuracy score is only ~70%, which isn't stellar. It's likely that they provide much of the signal to be had, but that a combined approach with further feature engineering would be helpful. 

Next steps:
- use further non-parametric methods on data as-is (XGBoost, SVM)
- feature engineer to factor out counties which may be dominating the model with much higher industry presence than others
- bring in outside relevant data to refine the AQI targets, this problem has more dimensions than good/bad (ex: temperature)
- model full YoY for TRI facilities to match YoY targets in AQI scores. 
- county level aggregation is a blunt object, there are more spohisticated methods to do geospatial analysis

Final thoughts:
Predicting air quality is complex. Weather and geography compund human activity (here, pollution) in a way that must be accounted for in a sophisticated model. Still, there seems to be clear signal between TRI pollutants and poor air quality. It's something which can easily be extraolpated out to other environmental metrics like water quality and rates of environmentally-induced disease or incidents (asthma, chemical fires).
